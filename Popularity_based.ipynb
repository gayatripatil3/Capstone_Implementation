```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df['published_date'] = pd.to_datetime(df['published_date'], errors='coerce')
df['Year'] = pd.to_datetime(df['published_date']).dt.to_period('Y')
```

```python
df2 = df.drop_duplicates(keep=False)
df2 = df2.drop_duplicates(['title'], keep='first')
```

```python
#replace NaN with an empty string
df2['description'] = df2['description'].fillna('')
#removing commas
df2['voters'] = df2['voters'].replace(',','', regex=True)
#filling with 0
df2['voters'] = df2['voters'].fillna(0)
#changing type for int
df2['voters'] = df2['voters'].astype(int)
df2.dtypes
```

```python
num_rating_df=df.groupby('title').count()['rating'].reset_index()
num_rating_df.rename(columns={'rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df=df.groupby('title').mean()['rating'].reset_index()
avg_rating_df.rename(columns={'rating':'avg_ratings'},inplace=True)
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on='title')
popular_df
```

```python
popular_df[popular_df['avg_ratings']>=3.5].sort_values('avg_ratings',ascending=False)
```

```python
top_6_books = popular_df[popular_df['avg_ratings'] >= 3.5].sort_values('avg_ratings', ascending=False).head(6)
```
